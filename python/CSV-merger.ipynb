```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('../data/band1_data_train.csv')
df2 = pd.read_csv('../data/band2_data_train.csv')
train = pd.read_json("../data/train.json")
```

```python
train.shape
```

```python
df1.columns = ['id', 'band_1_max', 'band_1_mean', 'band_1_variance', 'band_1_numobj']
df2.columns = ['id', 'band_2_max', 'band_2_mean', 'band_2_variance', 'band_2_numobj']
```

```python
df1.head()
```

```python
train = train.merge(df1, on='id', how='inner')
train = train.merge(df2, on='id', how='inner')
```

```python
train.head()
```

```python
train.to_json("../data/train_plus.json")
```
